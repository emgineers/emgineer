```python
# perform edge detection using canny edge detector on the image
# and display the result

import cv2
import numpy as np
img = cv2.imread('img.jpg')

edge_detect = cv2.Canny(img, 180, 200)
edge_detect = cv2.threshold(edge_detect, 127, 255, cv2.THRESH_BINARY)[1]

plt.imshow(edge_detect, cmap='gray')

contours, hierarchy = cv2.findContours(edge_detect, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(img, contours, -1, (0, 255, 0), 3)

blank_img = np.zeros(img.shape)
cv2.drawContours(blank_img, contours, -1, (0, 255, 0), 3)


plt.imshow(blank_img)

with open("path3.svg", "w+") as f:
    f.write(f'<svg width="{5930}" height="{3211}" xmlns="http://www.w3.org/2000/svg">')

    for c in contours:
        f.write('<path d="M')
        for i in range(len(c)):
            x, y = c[i][0]
            f.write(f"{x} {y} ")
        f.write('" style="stroke:pink"/>')
    f.write("</svg>")
```
